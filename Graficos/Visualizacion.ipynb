# Trends

Supongamos que queremos ver si el promedio de los sueldos varian por año. La variable tiempo que vamos a considerar es Joined que es la fecha cuando el jugador se unió a un club, y para ver el sueldo promedio deberíamos promediar por cada año la variable wage. 

```python
df['Joined'] = pd.to_datetime(df['Joined'])
df = df.sort_values(by = 'Joined')
df.head()
```

Si observamos la variable wage vemos que no es númerica, por lo tanto, no vamos a poder obtener un promedio. Deberíamos eliminar todo lo NO númerico. 

```python
df.Wage.unique()
```

```python
## quitamos los simbolos para trasnformarlas en numerica y poder multiplcarla
##wage hay que multiplicarla por mil que es lo que representa el K.
df['Wage'] = df.Wage.replace('\€|K', '', regex=True)
df['Wage'] = df['Wage'].astype('int')
df['Wage'] = df['Wage'] * 1000
```

```python
df.Wage.unique()
```

 Para ver el promedio de los salarios por año, tenemos que armar una variable del año de la fecha que se unio el jugador y después sacar promedio por año.

```python
df['year_joined'] = df['Joined'].dt.year
# hay años con Nan esos los sacamos
df = df[~(df.year_joined.isna())]
df1 = df[df.Club.isin([])]
df1 = df.groupby('year_joined').agg({'Wage':'mean'}).reset_index()
sns.lineplot(data=df1, x="year_joined", y="Wage")
```

Qué pasó en el 2004 que hay ese pico? 

Supongamos que ahora queremos lo mismo pero abierto por club. (Seleccionamos algunos porque al ser muchos clubs el gráfico no queda "bonito")

```python

df1 = df[df.Club.isin(['FC Barcelona', 'Real Madrid', 'Leicester City', 'Napoli', 'Arsenal'])]
df1 = df1.groupby(['year_joined','Club']).agg({'Wage':'mean'}).reset_index()
sns.lineplot(data=df1, x="year_joined", y="Wage", hue="Club")
```
